## Exercise 1

```python
from matplotlib import pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y,
xerr=x_err, yerr=y_err,
color='red', linestyle='--', # Example color, linestyle
marker='o', markerfacecolor='yellow', # Example marker, markerfacecolor
ecolor='black', capsize=5,
label='Y values with Errors') # Example label
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right') # Example legend location
plt.show()
```

## Exercise 2

```python
import numpy as np
```

```python
# Generate x values
x = np.linspace(0, 50, 500)

# Generate y values based on the given function
y = np.exp(-x / 10) * np.sin(x)

# Customize the plot
plt.plot(x, y, label=r'$e^{-\frac{x}{10}} \sin(x)$')  # Using LaTeX for the equation

plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title('Graph of $e^{-\\frac{x}{10}}\\sin(x)$', fontsize=20)

plt.grid(alpha=0.25)

plt.ylim([-1, 1])  # Set y-axis limit

plt.legend()
plt.show()
```
